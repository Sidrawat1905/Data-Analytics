```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("population.csv")
print(df.head())
print(df.columns)
```

```python
year = "2022"
top10 = df[["Country Name",year]].nlargest(10,year)

plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"],top10[year]/1e6, color = "skyblue", edgecolor = "black")
plt.show()
```

```python

countries_only = df[(df["Indicator Name"] == "Population, total")]

exclude_list = ["World", "High income", "Low income", "Upper middle income",
                "Lower middle income", "Middle income", "Low & middle income",
                "East Asia & Pacific", "Europe & Central Asia",
                "North America", "Latin America & Caribbean",
                "Sub-Saharan Africa", "South Asia", "Arab World"]

countries_only = countries_only[~countries_only["Country Name"].isin(exclude_list)]

year = "2022"
top10 = countries_only[["Country Name", year]].nlargest(10, year)

# Plot
plt.figure(figsize=(12,7))
bars = plt.bar(top10["Country Name"], top10[year]/1e6, color="skyblue", edgecolor="black")

plt.xlabel("Country")
plt.ylabel("Population (millions)")
plt.title(f"Top 10 Most Populated Countries in {year}")
plt.xticks(rotation=45, ha="right")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f"{height:.0f}", 
             ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

top10
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: load dataset
# df = pd.read_csv("population.csv")

# Select India row
india = df[df["Country Name"] == "India"].iloc[0]

# Extract years and population
years = india.index[4:].astype(int)             # skip metadata columns
values = pd.to_numeric(india.values[4:], errors="coerce")

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(years, values / 1e6, color="pink", linewidth=2, marker="o")

plt.xlabel("Year")
plt.ylabel("Population of India (millions)")
plt.title("Population Growth of India (1960 - 2022)")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)

plt.show()
```
